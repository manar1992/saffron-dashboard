```python
# 2. فك الضغط عن الملف
import zipfile
with zipfile.ZipFile("THZ_dataset_det_VOC.zip", 'r') as zip_ref:
    zip_ref.extractall("THZ_dataset_det_VOC")
```

```python
# 3. استيراد المكتبات
import os
import pandas as pd
import xml.etree.ElementTree as ET
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 4. تحليل البيانات من ملفات XML
annotations_path = "/content/THZ_dataset_det_VOC/Annotations"
images_path = "/content/THZ_dataset_det_VOC/JPEGImages"

# قراءة ملفات XML
annotations = []
for file in os.listdir(annotations_path):
    if file.endswith(".xml"):
        tree = ET.parse(os.path.join(annotations_path, file))
        root = tree.getroot()
        filename = root.find("filename").text
        labels = [obj.find("name").text for obj in root.findall("object")]
        for label in labels:
            annotations.append({"filename": filename, "label": label})
```

```python
# 5. تحويل البيانات إلى جدول
df = pd.DataFrame(annotations)

# طباعة معلومات عامة
print("Number of images:", df['filename'].nunique())
print("Number of classified weapons:", len(df))
print("\nFirst 5 rows:\n", df.head())
```

```python
# 6. رسم Countplot
plt.figure(figsize=(8, 5))
sns.countplot(x="label", data=df)
plt.title("Number of images for each weapon type")
plt.xlabel("Weapon type")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 7. رسم Histogram للمساحة
df['label'].value_counts().plot(kind='hist', bins=len(df['label'].unique()), edgecolor='black')
plt.title("Histogram of weapon types frequency")
plt.xlabel("Number of images")
plt.ylabel("Number of types")
plt.show()
```
